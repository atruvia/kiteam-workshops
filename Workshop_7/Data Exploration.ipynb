## Explore Variables

### Bar Plots

```python
f, ax = plt.subplots(figsize=(15, 6))
df_train['kills'].value_counts().sort_values(ascending=False).plot.bar()
plt.show()
```

### Histplots

```python
f, ax = plt.subplots(figsize=(15, 6))
sns.distplot(df_train['damageDealt'])
plt.show()
```

### MatchType

```python
f, ax = plt.subplots(figsize=(15, 6))
df_train['matchType'].value_counts().sort_values(ascending=False).plot.bar()
plt.show()
```

## Look for Correlation

### Correlation Matrix Top 10

```python
k = 10
corrmat = df_train.corr() 
cols = corrmat.nlargest(k, 'winPlacePerc').index 
cm = np.corrcoef(df_train[cols].values.T) 
sns.set(font_scale=1.25)
f, ax = plt.subplots(figsize=(16, 12))
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 12}, 
                 yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

### Scatter Plots

```python
df_train.plot(x="walkDistance",y="winPlacePerc", kind="scatter", figsize = (15,12))
plt.show()
```

```python
sns.jointplot(x="winPlacePerc", y="kills", data=df_train, ratio=3, color="r")
plt.show()
```
